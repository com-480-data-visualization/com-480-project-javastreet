```python
import pandas as pd
entries = pd.read_json('data/bigjson.txt')
entries = entries[['name','category', 'installs']]
entries.head(2)
```

```python
final_df = pd.DataFrame (columns = ['name', 'value', 'year', 'lastValue', 'category'])

for _, row in entries.iterrows():
    not_wanted=['Google', 'Android', 'Samsung', 'Gmail']
    
    if not any(x in row['name'] for x in not_wanted):
        lastCount = 0
        lastRoundedYear = 0
        formerRow = []
        
        for history in reversed(row['installs']['history']):
            roundedYear = "{:.1f}".format(int(history['date'][:4]) + float(history['date'][5:7])/12)
            currentCount = history['count']
            if lastRoundedYear != 0:
                nb_row = int(10*(float(roundedYear)-float(lastRoundedYear)))
                if nb_row:
                    incrementCount = (int(currentCount)-int(lastCount))/nb_row
                    #Makes sure all the interval between two milestones is completely filled
                    for i in range(1,nb_row+2):
                        final_df.loc[final_df.shape[0]] = formerRow
                        formerRow[3] = formerRow[1]
                        formerRow[1] = str(round(int(lastCount)+i*incrementCount))
                        formerRow[2] = "{:.1f}".format(float(lastRoundedYear)+i*0.1)

                    formerRow = [row['name']] + [currentCount] + [roundedYear] + [formerRow[3]] + [row['category']]
            else:
                formerRow = [row['name']] + [currentCount] + [roundedYear] + [lastCount] + [row['category']]
            lastRoundedYear = roundedYear
            lastCount = currentCount
        
        roundedYear = '2020.3'
        currentCount = row['installs']['current']['count']
        nb_row = int(10*(float(roundedYear)-float(lastRoundedYear)))
        if nb_row:
            incrementCount = (int(currentCount)-int(lastCount))/nb_row
            for i in range(1,nb_row+2):
                final_df.loc[final_df.shape[0]] = formerRow
                formerRow[3] = formerRow[1]
                formerRow[1] = str(round(int(lastCount)+i*incrementCount))
                formerRow[2] = "{:.1f}".format(float(lastRoundedYear)+i*0.1)
        final_df.loc[final_df.shape[0]] = [row['name']] + [currentCount] + [roundedYear] + [formerRow[3]] + [row['category']]
```

```python
#Removes any double entry
lastYear = '0000'
for i, row in final_df.iterrows():
    thisYear = row['year']
    if (thisYear==lastYear):
        final_df = final_df.drop(i)
    lastYear = thisYear
```
